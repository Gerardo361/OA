# Question 1. What is the Likelihood that we achieve a 15% engagement rate?
## Since every platform is different, its best to find their engagement rate separately.

```python
Twitter_social=df[df['Account Type']=='TWITTER']
Twitter_social.head()
```

```python
Twitter_percent= (Twitter_social['Total Engagements']/Twitter_social['Total Impressions']).dropna()#drops NaN values
print(Twitter_percent)
print(Twitter_percent[Twitter_percent > 0.15].count())
print(Twitter_percent[Twitter_percent > 0.15].count()/len(Twitter_percent))
```

## Around 1.8% of all Twitter engagment was over 15%.
